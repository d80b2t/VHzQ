```python
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt

from astropy.io import fits
```

```python
df = pd.read_csv('BestVSA_magnitudes_byRA.csv')
```

```python
len(df)
```

```python
for ii in range(len(df)-1):
    if ( df['filterID'][ii] == df['filterID'][ii+1] and df['qsoName'][ii] == df['qsoName'][ii+1] ):
        delta_magone = (abs(df['aperMag3AverAB'][ii]-df['aperMag3AverAB'][ii+1])/df['aperMag3Err'][ii])
        delta_magtwo = (abs(df['aperMag3AverAB'][ii]-df['aperMag3AverAB'][ii+1])/df['aperMag3Err'][ii+1])
        if((delta_magone > 5. and delta_magone <10000) and (delta_magtwo > 5. and delta_magtwo  <10000)):
        #if((delta_magone > 10000) or (delta_magtwo > 10000)):
            print(df['qsoName'][ii], 
                  df['filterID'][ii], 
                  df['aperMag3AverAB'][ii], 
                  delta_magone, delta_magtwo)
```

```python
dff = df.pivot_table(index='ra',
                     columns='filterID', 
                     values='aperMag3AverAB', 
                     fill_value='000.0000', 
                     aggfunc='mean')
```
